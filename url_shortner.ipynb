```python
import pandas as pd
from pandas import DataFrame
from json import loads
import numpy as np
```

```python
path = 'url-shortner/usagov_bitly_data2012-03-16-1331923249.txt'
```

```python
file = open(path, mode='r')
lines = []
for line in file:
    obj = loads(line)
    lines.append(obj)
```

```python
data = DataFrame(lines)
```

```python
s1 = data['tz']
s1.unique()
```

```python
s2 = s1[(s1.isnull()) | (s1 == '')]
s2
```

```python
s3 = s1.fillna('Missing')
s3
```

```python
s3[(s3.isnull()) | (s3 == '')]
```

```python
s3[s3 == ''] = 'Missing'
s3
```

```python
s3[(s3.isnull()) | (s3 == '')]
```

```python
data['tz'] = s3
data.head(20)
```

```python
# top 10 timezones using a lot of url shortening
d1 = data['tz']
d1
```

```python
d1.value_counts().head(10).plot(kind='bar')
```

```python
# top 10 timezones using a lot of url shortening, grouped OS wise (Windows/Nonwindows)
```

```python
data['os'] = np.where(data['a'].str.contains('Windows'), 'Windows', 'Non-Windows')
```

```python
data[['a', 'os']]
```

```python
d2 = data[['tz', 'os']]
d2
```

```python
d2['total'] = 1
d2
```

```python
d3 = d2.pivot_table('total', aggfunc='sum', index='tz', columns=['os']).fillna(0)
d3
```

```python
top_10_tzs = d1.value_counts().head(10).index
top_10_tzs
```

```python
d4 = d3.loc[top_10_tzs]
d4
```

```python
d4.plot(kind='bar')
```
